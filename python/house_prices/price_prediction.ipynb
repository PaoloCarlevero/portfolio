## Mutual Information

```python
def make_mi_scores(data, target):

    X = data.copy()
    y = X.pop(target)

    mi_scores = {}
    for column in X.columns:
        if X[column].dtype in ('int32', 'int64', 'float32', 'float64'):
            column_values = X[column].dropna()
            target_values = y.loc[X.index.isin(column_values.index)]

            column_mi_score = mutual_info_regression(np.array(column_values).reshape(-1,1), target_values)
            mi_scores[column] = round(column_mi_score[0], 4)

    return pd.Series(mi_scores, dtype='Float64')

mi_scores = make_mi_scores(train, 'SalePrice')
```

```python
def plot_mi_score(mi_scores):
    mi_scores = mi_scores.sort_values(ascending=False)
    sns.barplot(y=mi_scores.index, x=mi_scores.values, color='b')
    plt.title("Mutual Information Scores")

plot_mi_score(mi_scores)
```

```python
def break_down_MSSubClass(item):
    if item in [20, 30, 40, 120]:
        return '1-STORY'
    elif item in [45, 50, 150]:
        return '1-1/2 STORY'
    elif item in [60, 70, 160]:
        return '2-STORY'
    elif item in [75]:
        return '2-1/2 STORY'
    elif item in [80, 85, 180]:
        return 'SPLIT'
    elif item in [90, 190]:
        return 'DUPLEX'
    else:
        raise Exception(f"Unrecognised value: {item}")

train['MSSubClass'] = train['MSSubClass'].apply(break_down_MSSubClass)
```
