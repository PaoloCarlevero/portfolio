#### Checking the number of records for each date

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
def get_records_per_date(df):
    return df.groupby(by='date')['location'].count()

records_per_date = get_records_per_date(df)
records_per_date
```

```python
ax = plt.subplot()

ax.bar(x=records_per_date.keys(), height=records_per_date.values)
plt.xticks(rotation=45)
plt.show()
```

#### 2020-01-01 and 2020-01-02 are not supposed to be tracked, since the dataset starts from 2020-01-03

```python
df.loc[df['date'] < pd.Timestamp(2020, 1, 3)]
```

All the records regarding 2020-01-01 and 2020-01-02 do not contain any covid related infroamtion, so they can be dropped

```python
df.drop(df[ ( df['date'] < pd.Timestamp(2020, 1, 3) ) | ( df['date'] > pd.Timestamp.today() )].index, inplace = True)
```

#### We continue to check the other dates

```python
def get_dates_per_num_of_records(df):
    dates_per_record = df.groupby(by='date')['location'].count().values
    unique, counts = np.unique(dates_per_record, return_counts= True)

    return dict(zip(unique, counts))

get_dates_per_num_of_records(df)
```

#### Getting the list of all locations

```python
locations_list = list(df['location'].unique())

locations_list
```

```python
current_date = df['date'].min()
end_date  = df['date'].max()
index = df.index.max() + 1

while current_date <= end_date:

    slice = df.loc[df['date'] == current_date]
    slice_locations = set(slice['location'])

    for location in locations_list:
        if location not in slice_locations:
            df.loc[index] = {'location': location, 'date': current_date}
            index += 1

    current_date = current_date + timedelta(days=1)
```

```python
get_dates_per_num_of_records(df)
```

```python
def get_records_per_date(df):
    return df.groupby(by='date')['location'].count()

records_per_date = get_records_per_date(df)
records_per_date
```

```python
ax = plt.subplot()

ax.bar(x=records_per_date.keys(), height=records_per_date.values)
plt.xticks(rotation=45)
plt.show()
```

#### As we can see, all dates now have the same amount of records.
#### We still need to sort the rows and reset the indexes.

```python
df = df.sort_values(['location_type', 'continent', 'location', 'date']).reset_index(drop=True)

df
```

The newly added rows have most values as np.NaN, we need to fill them

```python
df['location_type'] = np.NaN
```

```python
def get_location_type(item):

    if item in ('Africa', 'Asia', 'Europe', 'South America', 'North America', 'Oceania'):
        return 'continent'
    elif item in ('European Union'):
        return 'economic_uninon'
    elif item in ('High income', 'Low income', 'Lower middle income', 'Upper middle income'):
        return 'income'
    elif item == 'World':
        return 'world'
    else:
        return 'country'

df['location_type'] = df['location'].apply(get_location_type)
```

#### Filling np.NaN in continent column

```python
df.loc[df['location_type'] == 'continent', 'continent'] = df.loc[df['location_type'] == 'continent', 'location']
```

```python
df.loc[df['location_type'] == 'economic_uninon', 'continent'] = df.loc[df['location_type'] == 'economic_uninon', 'location']
```

```python
df.loc[df['location_type'] == 'income', 'continent'] = df.loc[df['location_type'] == 'income', 'location']
```

```python
df.loc[df['location_type'] == 'world', 'continent'] = 'World'
```
