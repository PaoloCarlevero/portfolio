## Handling NaN

### Checking how many NaN there are and the dtype of the column they belong

```python
nan_series = train.isna().sum().loc[lambda x: x > 0]

nan_df = pd.DataFrame({
        'column': nan_series.index,
        'num_of_nan': nan_series.values,
        'dtype': train[nan_series.index].dtypes.values
})

nan_df.sort_values(by='dtype')
```

## Year columns

### Converting column who express a year in into datetime format

```python
year_columns = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']

pd.to_datetime(train['GarageYrBlt'], format='%Y')

train[year_columns] = train[year_columns].apply(pd.to_datetime, axis=1, format='%Y')
train[year_columns]
```

### Checking if ther is a relationship between the time when tho housa ha ben sold and the price

```python
train['YrMoSold'] = train['YrSold'].astype(str) + "-" + train['MoSold'].astype(str)
train['YrMoSold'] = pd.to_datetime(train['YrMoSold'], format='%Y-%m')
train['YrMoSold']
```

```python
sns.relplot(data=train, x='YrMoSold', y='SalePrice')
plt.xticks(rotation=45)
plt.show()
```

Since there are no relationship between the time when a house has ben sold and the price 'MoSold','YrSold', 'YrMoSold' can be dropped. Also this information wouldn't be available when evaluating the selling price of a house.

```python
train.drop(labels=['MoSold','YrSold', 'YrMoSold'], axis=1, inplace=True)
```
