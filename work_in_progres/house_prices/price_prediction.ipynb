## Datetime columns

### Converting column who express a year in into datetime format

```python
year_columns = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']

pd.to_datetime(train['GarageYrBlt'], format='%Y')

train[year_columns] = train[year_columns].apply(pd.to_datetime, axis=1, format='%Y')
train[year_columns]
```

### Checking if there is a relationship between the time when a house has been sold and its price

#### Joining YrSOld and MoSold to get a more accurate selling date

```python
train['YrMoSold'] = train['YrSold'].astype(str) + "-" + train['MoSold'].astype(str)
train['YrMoSold'] = pd.to_datetime(train['YrMoSold'], format='%Y-%m')
train['YrMoSold']
```

#### Coverting YrMoSold in a distance from the first house  that has been sold expressed in days since datetime isn't recognised by (most) regression algorithm

```python
min_day = train['YrMoSold'].min()
train['TimeDelta'] = train['YrMoSold'].apply(lambda x: (x - min_day).days)

sns.regplot(data=train, x='TimeDelta', y='SalePrice', line_kws={'color':'red'})
plt.xticks(rotation=45)
plt.show()
```

```python
model = LinearRegression()
model.fit(train.loc[:, ['TimeDelta']], train.loc[:, 'SalePrice'])

model.score(train.loc[:, ['TimeDelta']], train.loc[:, 'SalePrice'])
```

Since there are no relationship between the time when a house has ben sold and the price 'MoSold','YrSold', 'YrMoSold' can be dropped. Also this information wouldn't be available when evaluating the selling price of a house.

```python
train.drop(labels=['MoSold','YrSold', 'YrMoSold', 'TimeDelta'], axis=1, inplace=True)
```
