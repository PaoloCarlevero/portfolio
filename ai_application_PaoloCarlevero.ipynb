```python
import fitz #PyMuPDF
import transformers

from transformers import pipeline
```

# Models set-up

```python
bert_qa = pipeline("question-answering", model="bert-large-uncased-whole-word-masking-finetuned-squad")
distilbert_qa = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")
```

# Importing the Articles text

```python
article_7 = ""
with fitz.open("Article 7 Efficient Estimation of Word Representations in Vector Space.pdf") as article:
  for page_num in range(len(article)):
    page = article[page_num]
    article_7 += page.get_text()

article_7
```

```python
article_8 = ""
with fitz.open("Article 8 Llama 2_ Open Foundation and Fine-Tuned Chat Models.pdf") as article:
  for page_num in range(len(article)):
    page = article[page_num]
    article_8 += page.get_text()

article_8
```

# Custom questions

```python
# Article 7
article_7_question_1 = "How are words currently tretaed in NLP systems?"
article_7_question_2 = "Which are the model architectures are used to minimze computanional complexity?"
article_7_question_3 = "How is the word vectors' quality measured?"
article_7_question_4 = "Give examples of Learned Relationship"
article_7_question_5 = "Is it possible to compute high-quality word vectors using simple model architectures?"

# Article 8
article_8_question_1 = "What is fine-tuning, and which techniques can be used?"
article_8_question_2 = "How LLMs can be avaluated?"
article_8_question_3 = "Which safety measure can be taken during Pretraining?"
article_8_question_4 = "What is Reward Modelling?"
article_8_question_5 = "What is False Refusal?"
```

## Question answering

## Article 7

```python
for question in [article_7_question_1, article_7_question_2, article_7_question_3, article_7_question_4, article_7_question_5]:
  print(question)
  print(distilbert_qa(question=question, context=article_7))
  print("\n")
```

```python
for question in [article_7_question_1, article_7_question_2, article_7_question_3, article_7_question_4, article_7_question_5]:
  print(question)
  print(bert_qa(question=question, context=article_7))
  print("\n")
```
